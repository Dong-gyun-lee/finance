### 3. Intrinio에서 데이터 불러오기

먼저 https://intrinio.com 에 등록하여 개별 api key를 먼저 받아야 한다.

```python
import intrinio_sdk
import pandas as pd
```

```python
# key는 개인 api key 중 sandbox key 사용
intrinio_sdk.ApiClient().configuration.api_key['api_key'] = '개별 API KEY'
security_api = intrinio_sdk.SecurityApi()
```

```python
r = security_api.get_security_stock_prices(identifier='AAPL',
                                           start_date='2020-01-01',end_date='2021-12-30',frequency='daily',page_size=1000)
```

```python
response_list = [x.to_dict() for x in r.stock_prices]
df_intrinio = pd.DataFrame(response_list).sort_values('date')
df_intrinio.set_index('date',inplace=True)
```
